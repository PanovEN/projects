## Цель исследования:

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.  

### Описание услуг

Оператор предоставляет два основных типа услуг: 
1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*). 

Также доступны такие услуги:
- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).  

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.  

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract_new.csv` — информация о договоре;
- `personal_new.csv` — персональные данные клиента;
- `internet_new.csv` — информация об интернет-услугах;
- `phone_new.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

```python
# Создадим целевой признак outflow - отток клиента.
# Если в EndDate - "No", значит клиент действующий (0), иначе клиент ушёл (1):
def sign(data):
    if data == 'No':
        result = 0
    else:
        result = 1
    return result

contract['outflow'] = contract.EndDate.apply(sign)
```

```python
# Заменим "No" в EndDate на дату актуальности данных(01.02.20), чтобы подсчитать
# количества дней работы  с клиентом
def end_date(data):
    if data == 'No':
        result = '2020-02-01'
    else:
        result = data
    return result

contract['EndDate'] = contract.EndDate.apply(end_date)
```

```python
# Преобразуем "BeginDate" и 'EndDate' в формат даты:
contract['BeginDate'] = pd.to_datetime(contract['BeginDate'], format='%Y-%m-%d')
contract['EndDate'] = pd.to_datetime(contract['EndDate'], format='%Y-%m-%d')
```

```python
# Рассчитаем количество дней сотрудничества
contract['active_days'] = (contract['EndDate'] - contract['BeginDate']).astype(
    'timedelta64[D]').astype('int')
contract.head()
```

```python
# Преобразуем категориальные признаки в численные с помощью OneHotEncoder
encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
encoder_ohe.fit(X_train[categorical_features])
X_train_ohe = X_train.copy()
X_train_ohe[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(
    X_train[categorical_features])
X_train_ohe = X_train_ohe.drop(categorical_features, axis=1)
X_train_ohe.head()
```

```python
# То же самое сделаем с тестовой выборкой
X_test_ohe = X_test.copy()
X_test_ohe[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(
    X_test[categorical_features])
X_test_ohe = X_test_ohe.drop(categorical_features, axis=1)
X_test_ohe.head()
```

```python
# Отмасштабируем количественные признаки
numeric = ['MonthlyCharges', 'TotalCharges', 'active_days']
scaler = StandardScaler()
scaler.fit(X_train_ohe[numeric])

X_train_scaled = X_train_ohe.copy()
X_test_scaled = X_test_ohe.copy()

X_train_scaled[numeric] = scaler.transform(X_train_scaled[numeric])
X_test_scaled[numeric] = scaler.transform(X_test_scaled[numeric])
X_train_scaled.head()
```

**Данные готовы для обучения моделей**

```python
# Функция для обучения моделей, с кросс-вадидацией
def model_training(model, param):
    cv = KFold(shuffle=True, random_state=random_state)
    gsearch_model = GridSearchCV(model, param_grid=param, n_jobs=-1, scoring='roc_auc', cv=cv)
    gsearch_model.fit(X_train_scaled, Y_train)
    print('Лучшая ROC-AUC: ', gsearch_model.best_score_)
    print('Лучшие параметры модели: ', gsearch_model.best_params_)
    print('Лучшая модель: ', gsearch_model.best_estimator_)
    winsound.Beep(440, 500)
    return gsearch_model.best_estimator_, round(gsearch_model.best_score_, 2)
```

**LightGBM**

```python
%%time
model_lgbm = LGBMClassifier(random_state=random_state, n_jobs=-1, is_unbalance=True,
                            force_col_wise=True)
param_lgbm = {
    'learning_rate' : np.arange(0.42, 0.44, 0.01),
    'boosting_type' : ['gbrt', 'goss', 'dart'],
    'max_depth' : range(3,6)
    }

best_lgbm, lgbm_score = model_training(model_lgbm, param_lgbm)
```

ROC-AUC=0.89, лучшие параметры: 'boosting_type': 'gbrt', 'learning_rate': 0.43, 'max_depth': 4. Время настройки параметров 3,23сек.

**CatBoost**

```python
%%time
model_catboost = CatBoostClassifier(random_state=random_state, verbose=0)
param_catboost = {'n_estimators': range(594, 597),
                  'learning_rate': np.arange(0.34, 0.36, 0.01),
                  'max_depth': range(1, 4)
                  }

best_catboost, catboost_score = model_training(model_catboost, param_catboost)
```

ROC-AUC=0.93, лучшие параметры: 'n_estimators': 595, 'learning_rate': 0.35, 'max_depth': 2. Время настройки параметров 30,8сек.

**XGBoost**

```python
%%time
model_xgboost = XGBClassifier(random_state=random_state, n_jobs=-1)
param_xgboost = {'n_estimators': range(857, 860),
                 'learning_rate': np.arange(0.2, 0.22, 0.01),
                 'max_depth': range(1, 4),
                }

best_xgboost, xgboost_score = model_training(model_xgboost, param_xgboost)
```
